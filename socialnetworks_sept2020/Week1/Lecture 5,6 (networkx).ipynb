```python
# h is a complete graph i.e each node connected to all other nodes
h=nx.complete_graph(10)
print('Nodes: ',h.nodes(),'\n')
print('Edges: ',h.edges(),'\n')
print('No. of nodes: ',h.order())
print('No. of edges: ',h.size())
nx.draw(h,with_labels=1)
```

Lecture 6 : Networkx II

```python
import networkx as nx
h=nx.Graph()           #undirected graph
                       #g=nx.DiGraph()   Diracted graph
city=['Delhi','Banglore','Hyderabad','Ahmedabad','Chennai','Kolkata','Surat','Pune']
for each in city:
    h.add_node(each)
nx.draw(h,with_labels=1)
h.nodes()
```

```python
cost=[]
value=100
while(value<=1000):
    cost.append(value)
    value+=100
cost
```

```python
#adding 16 edges
import random
while(h.number_of_edges()<16):
    c1=random.choice(city)
    c2=random.choice(city)
    if c1!=c2 and h.has_edge(c1,c2)==False:
        w=random.choice(cost)
        h.add_edge(c1,c2,weight=w)
nx.draw(h,with_labels=1)
```

```python
#Changing the graph layout
pos=nx.circular_layout(h)
nx.draw(h,pos,with_labels=1)
#or  nx.draw_circular(h,with_labels=1) 
```

```python
#displaying the graph and weight to each node
pos=nx.circular_layout(h)
nx.draw_networkx_edge_labels(h,pos)
```

```python
nx.is_connected(h)
```

```python
#checking for a path from each a city to every other city
for u in g.nodes():
    for v in g.nodes():
        print (u,v,nx.has_path(g,u,v))
```

```python
#using dijkstra path finding algorithm to find the path and cross checking for the cost
u='Kolkata'
v='Pune'
print(nx.dijkstra_path(h,u,v))
print(' ',nx.dijkstra_path_length(h,'Kolkata','Hyderabad'))
print('+',nx.dijkstra_path_length(h,'Hyderabad','Banglore'))
print('+',nx.dijkstra_path_length(h,'Banglore','Pune'))
print('-------')
print(nx.dijkstra_path_length(h,u,v))
```

```python
#making new graph z
import networkx as nx
z=nx.Graph()
city=['Delhi','Banglore','Hyderabad','Ahmedabad','Chennai','Kolkata','Surat','Pune']
for i in city:
    z.add_node(i)
nx.draw(z,with_labels=1)
```

```python
#since the graph is having no edges and there's no path between 2 cities dijkstras will throw a exception and except block will run
try:
    l=nx.dijkstra_path_length(z,'Delhi','Banglore')
except:
    l=5000
print('Path length: ',l)
```

```python
import random
import matplotlib.pyplot as plt
cost=[100,200,300,400,500,600,700,800,900,1000]
def add_random_edge(z,cost):
    c1=random.choice(city)
    c2=random.choice(city)
    if c1!=c2 and z.has_edge(c1,c2)==False:
        w=random.choice(cost)
        z.add_edge(c1,c2,weight=w)
x=[0]
y=[l]
#trying to add edges 10 time and the edges made can be <=10 because of the if condtn in add_random_edge()
for i in range(1,11):
    add_random_edge(z,cost)
    x.append(i)
    try:
        l=nx.dijkstra_path_length(z,'Delhi','Banglore')
        print(l)
        y.append(l)
        break
    except:
        l=5000
        y.append(l)
        print('No path')
print(x,y)
plt.xlabel('No. of Edges')
plt.ylabel('Travelling Cost')
plt.title('edges vs cost')
plt.plot(x,y)
```
